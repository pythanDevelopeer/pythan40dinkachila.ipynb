```python
df.name.str.split(' ').head()
```

```python
df.name.str.split(' ').expand=True
```

```python
# spliting a column into two columns
df.name.str.split(' ',expand=True)
```

```python
# Adding those split into new column
df[["first_name","last_name"]]=df.name.str.split(' ',expand=True)
df
```

```python
# scatter plot
plt.scatter(data=data_set, x='age', y='fare', c=data_set['pclass'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.colorbar(label='Pclass')
plt.show()
```

```python
sns.scatterplot(data=data_set,x='age',y="fare",hue="pclass")
```

```python
sns.barplot(data=data_set,x='sex',y="age",hue="pclass")
```

```python
sns.pairplot(data_set)
```

```python
sns.boxplot(data=data_set,x='sex',y="fare",hue="parch")
```

```python
sns.boxenplot(data=data_set,x='sex',y="fare",hue="parch");
```

```python
g=sns.FacetGrid(data_set,hue='survived',col='pclass',margin_titles=True,
               palette={1:'seagreen',0:'gray'})
g=g.map(plt.scatter,"fare","age",edgecolor='w').add_legend();
```

```python
g=sns.FacetGrid(data_set,hue='survived',col='sex',margin_titles=True,
               palette={1:'seagreen',0:'gray'})
g=g.map(plt.scatter,"fare","age",edgecolor='w').add_legend();
```

```python
g=sns.FacetGrid(data_set,hue='survived',col='who',margin_titles=True,
               palette={1:'seagreen',0:'gray'})
g=g.map(plt.scatter,"fare","age",edgecolor='w').add_legend();
```
